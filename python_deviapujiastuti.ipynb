```python
import pandas as pd
import plotly.express as px

# Load dataset
file_path = "/content/exams.csv"
df = pd.read_csv(file_path)

# Membuat histogram interaktif
fig = px.histogram(df,
                   x='math score',  # Kolom nilai matematika
                   color='gender',  # Warna berdasarkan gender
                   barmode='overlay',  # Overlay untuk membandingkan distribusi
                   nbins=20,  # Jumlah bin (kelompok nilai)
                   title='Distribusi Nilai Matematika Berdasarkan Gender',
                   labels={'math score': 'Nilai Matematika', 'gender': 'Jenis Kelamin'},
                   opacity=0.6  # Transparansi agar grafik lebih mudah dibaca
                   )

# Menampilkan grafik interaktif
fig.show()
```
